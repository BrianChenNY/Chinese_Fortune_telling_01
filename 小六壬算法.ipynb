```python
import datetime
from lunarcalendar import Converter, Solar, Lunar, DateNotExist
# from datetime import datetime
import pandas as pd
import numpy as np
import re
import os
import scipy as sp

# importing matplotlib modules 
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
def lunar_hour (hour):
    
        if hour in [23, 0]:
            lunar_hour = 1
        elif hour in [1, 2]:
            lunar_hour = 2
        elif hour in [3, 4]:
            lunar_hour = 3
        elif hour in [5, 6]:
            lunar_hour = 4
        elif hour in [7, 8]:
            lunar_hour = 5
        elif hour in [9, 10]:
            lunar_hour = 6
        elif hour in [11, 12]:
            lunar_hour = 7
        elif hour in [13, 14]:
            lunar_hour = 8
        elif hour in [15, 16]:
            lunar_hour = 9
        elif hour in [17, 18]:
            lunar_hour = 10
        elif hour in [19, 20]:
            lunar_hour = 11
        else:
            lunar_hour = 12
        
        return lunar_hour
```

```python
def SIXCHANGE (time, birth):
    try: 
        stages = {1:'   大安：气运正盛，加仓入市   ', 2:'   留连：蛰伏盘整，持仓待变   ',\
                  3:'   速喜：浑水摸鱼，短线操作   ', 4:'   赤口： 多空博弈，减仓为上  ', \
                  5:'   小吉：稳中看好，半仓试探   ', 0:'   空亡：十面埋伏，不可触碰   '} 

        time =datetime.datetime.strptime(time, '%Y-%m-%d-%H')
        
        birth =datetime.datetime.strptime(birth, '%Y-%m-%d-%H')

        solar = Solar(time.year, time.month, time.day)

        lunar = Converter.Solar2Lunar(solar)
        
        b_solar = Solar(time.year, time.month, time.day)

        b_lunar = Converter.Solar2Lunar(b_solar)


        sumup = lunar.month + lunar.day + b_lunar.month + b_lunar.day

        hour1 = time.hour
        
        hour2 = birth.hour
        
        lunar_hour1 = lunar_hour(hour1)
        
        lunar_hour2 = lunar_hour(hour2)
        
        
        sumup += lunar_hour1 + lunar_hour2 - 4

        num_result = sumup % 6
        
        if num_result == 1:
            
            plt.imshow(mpimg.imread('daan.jpg'));
        
        if num_result == 2:
            
            plt.imshow(mpimg.imread('liulian.jfif'));
                
        if num_result == 3:
            
            plt.imshow(mpimg.imread('suxi.png'));
                
        if num_result == 4:
            
            plt.imshow(mpimg.imread('chikou.jpg'));
                
        if num_result == 5:
            
            plt.imshow(mpimg.imread('xiaoji.png'));                
                
        if num_result == 0:
            
            plt.imshow(mpimg.imread('kongwan.jpg')); 
        

        return '\n\n'+ stages[num_result]
    
    except:
        
        return "Format Wrong"
    
```

```python
brithday = input(" 输入生辰四柱 (依此格式：yyyy-mm-dd-hh): ") 
timing = input(" 输入准备入场的时间 (依此格式：yyyy-mm-dd-hh): ") 

print(SIXCHANGE(timing, brithday)) 
```
